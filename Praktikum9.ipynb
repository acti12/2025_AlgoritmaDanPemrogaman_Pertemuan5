# Penugasan 1 - Eksperimen Dengan Modul Math

```python
import math

def luas_lingkaran(jari_jari):
    """Menghitung luas lingkaran menggunakan math.pi"""
    luas = math.pi * (jari_jari ** 2)
    return luas

# Uji dengan jari-jari 7 dan 10
r1 = 7
r2 = 10

luas1 = luas_lingkaran(r1)
luas2 = luas_lingkaran(r2)

print(f"Luas lingkaran dengan jari-jari {r1} = {luas1:.2f}")
print(f"Luas lingkaran dengan jari-jari {r2} = {luas2:.2f}")
print(f"\nNilai math.pi = {math.pi}")

print("\n")
```

# Penugasan 4 - Modul Buatan Sendiri

```python
konversi_code = '''# konversi.py
# Modul untuk konversi suhu

def celcius_ke_fahrenheit(c):
    """Konversi Celcius ke Fahrenheit"""
    f = (c * 9/5) + 32
    return f

def fahrenheit_ke_celcius(f):
    """Konversi Fahrenheit ke Celcius"""
    c = (f - 32) * 5/9
    return c
'''

# Simpan ke file konversi.py
with open('konversi.py', 'w') as f:
    f.write(konversi_code)

print("✓ File konversi.py berhasil dibuat!")
print("\nIsi konversi.py:")
print(konversi_code)
```

```python
import konversi

print("\n--- Penggunaan Modul konversi.py ---")

# Test konversi
suhu_c = 30
suhu_f = 86

hasil_f = konversi.celcius_ke_fahrenheit(suhu_c)
hasil_c = konversi.fahrenheit_ke_celcius(suhu_f)

print(f"{suhu_c}°C = {hasil_f:.2f}°F")
print(f"{suhu_f}°F = {hasil_c:.2f}°C")

# Test dengan nilai lain
print("\nTest tambahan:")
suhu_list_c = [0, 25, 37, 100]
for suhu in suhu_list_c:
    hasil = konversi.celcius_ke_fahrenheit(suhu)
    print(f"{suhu}°C = {hasil:.2f}°F")

print("\n")
```

# Penugasan 6 - Studi Kasus Mini

```python
laporan_code = '''# laporan.py
# Modul untuk menampilkan laporan kuis

def tampilkan_ringkasan(nama, total_soal, benar, salah, skor):
    """Menampilkan ringkasan hasil kuis"""
    print("\\n" + "="*50)
    print("          RINGKASAN HASIL KUIS")
    print("="*50)
    print(f"Nama Peserta    : {nama}")
    print(f"Total Soal      : {total_soal}")
    print(f"Jawaban Benar   : {benar}")
    print(f"Jawaban Salah   : {salah}")
    print(f"Skor Akhir      : {skor}/{total_soal*10}")
    print(f"Persentase      : {(skor/(total_soal*10))*100:.1f}%")
    print("="*50)

    # Tentukan grade
    persentase = (skor/(total_soal*10))*100
    if persentase >= 80:
        grade = "A - Sangat Baik!"
    elif persentase >= 70:
        grade = "B - Baik!"
    elif persentase >= 60:
        grade = "C - Cukup"
    else:
        grade = "D - Perlu Belajar Lagi"

    print(f"Grade           : {grade}")
    print("="*50)

def simpan_laporan(nama, skor, filename="hasil_kuis.txt"):
    """Menyimpan hasil kuis ke file"""
    with open(filename, 'a') as f:
        f.write(f"{nama}: {skor}\\n")
    print(f"\\n✓ Hasil tersimpan di {filename}")
'''

# Simpan ke file laporan.py
with open('laporan.py', 'w') as f:
    f.write(laporan_code)
```

```python
import laporan

print("\n" + "="*50)
print("          PROGRAM KUIS MATEMATIKA")
print("="*50)

# Input nama peserta
nama_peserta = input("\nMasukkan nama Anda: ")

# Konfigurasi kuis
jumlah_soal = 5
skor_per_soal = 10

# Variabel untuk tracking
jawaban_benar = 0
jawaban_salah = 0
total_skor = 0

print(f"\nKuis akan dimulai! Anda akan menjawab {jumlah_soal} soal penjumlahan.")
print("="*50)

# Generate soal dan proses jawaban
for i in range(jumlah_soal):
    # Generate dua angka acak
    angka1 = random.randint(1, 50)
    angka2 = random.randint(1, 50)
    jawaban_benar_soal = angka1 + angka2

    # Tampilkan soal
    print(f"\nSoal {i+1}:")
    print(f"{angka1} + {angka2} = ?")

    # Terima jawaban
    try:
        jawaban_user = int(input("Jawaban Anda: "))

        # Cek jawaban
        if jawaban_user == jawaban_benar_soal:
            print("✓ BENAR!")
            jawaban_benar += 1
            total_skor += skor_per_soal
        else:
            print(f"✗ SALAH! Jawaban yang benar: {jawaban_benar_soal}")
            jawaban_salah += 1
    except ValueError:
        print("✗ SALAH! Input tidak valid.")
        jawaban_salah += 1

# Tampilkan ringkasan menggunakan modul laporan
laporan.tampilkan_ringkasan(
    nama_peserta,
    jumlah_soal,
    jawaban_benar,
    jawaban_salah,
    total_skor
)

# Simpan hasil
laporan.simpan_laporan(nama_peserta, total_skor)

print("\n✓ Program kuis selesai!")
print("\n" + "="*60)
```
