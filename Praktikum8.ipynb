# Praktikum 4 - Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 7 - Fungsi Dekorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()

    print("Setelah fungsi dijalankan...")
  return wrapper
@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

# Penugasan 3 - Fungsi Dengan Default Arguments

```python
def salam(waktu="pagi", nama="Mahasiswa"):
    print(f"Selamat {waktu}, {nama}!")

# Panggil dengan berbagai kombinasi
print("Tanpa argumen:")
salam()

print("\nDengan 1 argumen (waktu):")
salam("siang")

print("\nDengan 2 argumen lengkap:")
salam("malam", "Budi")

print("\nDengan keyword argument:")
salam(nama="Siti")

print()
```

# Penugasan 4 - Fungsi dengan *args

```python
def jumlahkan_semua(*angka):
    total = sum(angka)
    return total

# Uji dengan 3, 5, dan 7 bilangan
hasil1 = jumlahkan_semua(1, 2, 3)
hasil2 = jumlahkan_semua(5, 10, 15, 20, 25)
hasil3 = jumlahkan_semua(2, 4, 6, 8, 10, 12, 14)

print(f"Jumlah dari 3 bilangan (1,2,3): {hasil1}")
print(f"Jumlah dari 5 bilangan (5,10,15,20,25): {hasil2}")
print(f"Jumlah dari 7 bilangan (2,4,6,8,10,12,14): {hasil3}")

print()
```

# Penugasan 5 - Fungsi dengan **kwargs

```python
def info_barang(**data):
    for kunci, nilai in data.items():
        print(f"{kunci}: {nilai}")

print("Contoh 1:")
info_barang(Nama="Buku", Harga=15000, Stok=30)

print("\nContoh 2:")
info_barang(Nama="Pensil", Harga=3000, Stok=100, Merk="Faber-Castell")

print()
```

# Penugasan 6 - Scope Variable

```python
x = 100  # Variabel global

def ubah():
    x = 50  # Variabel lokal
    print(f"Nilai x di dalam fungsi (lokal): {x}")

print(f"Nilai x sebelum fungsi dipanggil (global): {x}")
ubah()
print(f"Nilai x setelah fungsi dipanggil (global): {x}")
print()
```

# Penugasan 7 - Fungsi Lambda

```python
kuadrat = lambda x: x ** 2

print("a. Lambda untuk kuadrat:")
print(f"Kuadrat dari 5 = {kuadrat(5)}")
print(f"Kuadrat dari 8 = {kuadrat(8)}")

# b. Lambda untuk keliling lingkaran
import math
keliling_lingkaran = lambda r: 2 * math.pi * r

print("\nb. Lambda untuk keliling lingkaran:")
print(f"Keliling lingkaran dengan r=7 = {keliling_lingkaran(7):.2f}")
print(f"Keliling lingkaran dengan r=10 = {keliling_lingkaran(10):.2f}")

print()
```

# Penugasan 8 - Fungsi Decorator

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper

@logger
def halo():
    print("Halo, selamat datang!")

halo()

print()
```

# Penugasan 9 - Studi Kasus Nyata : Program Nilai

```python
def input_nilai():
    """Menerima input nilai mahasiswa"""
    nilai_list = []
    jumlah = int(input("Masukkan jumlah mahasiswa: "))
    for i in range(jumlah):
        nilai = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
        nilai_list.append(nilai)
    return nilai_list

def hitung_rata(nilai_list):
    """Menghitung rata-rata nilai"""
    if len(nilai_list) == 0:
        return 0
    return sum(nilai_list) / len(nilai_list)

def status_kelulusan(rata):
    """Menentukan status kelulusan"""
    if rata >= 60:
        return "LULUS"
    else:
        return "TIDAK LULUS"

# Lambda untuk nilai akhir berdasarkan bobot
nilai_akhir = lambda tugas, ujian: (tugas * 0.4) + (ujian * 0.6)

# Contoh penggunaan program
print("\n=== PROGRAM PENILAIAN MAHASISWA ===\n")

# Simulasi tanpa input (untuk demo)
print("--- Contoh Data ---")
daftar_nilai = [75, 80, 65, 90, 55]
print(f"Daftar nilai: {daftar_nilai}")

rata_rata = hitung_rata(daftar_nilai)
print(f"\nRata-rata nilai: {rata_rata:.2f}")

status = status_kelulusan(rata_rata)
print(f"Status: {status}")

# Contoh perhitungan nilai akhir dengan bobot
print("\n--- Contoh Perhitungan Nilai Akhir ---")
tugas_mahasiswa = 75
ujian_mahasiswa = 85
nilai_final = nilai_akhir(tugas_mahasiswa, ujian_mahasiswa)
print(f"Nilai Tugas: {tugas_mahasiswa}")
print(f"Nilai Ujian: {ujian_mahasiswa}")
print(f"Nilai Akhir (40% tugas + 60% ujian): {nilai_final:.2f}")
print(f"Status: {status_kelulusan(nilai_final)}")

print("\n" + "="*50)
print("PROGRAM SELESAI")
print("="*50)
```
